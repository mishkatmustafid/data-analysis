```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
data = df.copy()

# Encode IncGeo_WardName column using LabelEncoder
le = LabelEncoder()
data['IncGeo_WardName'] = le.fit_transform(data['IncGeo_WardName'])

# select relevant columns
cols = ['FirstPumpArriving_AttendanceTime', 'IncGeo_WardName', 'IncidentGroup', 'DatetimeOfCall', 'FirstPumpArriving_DeployedFromStation']
data = data[cols]

# create dummy variables for categorical columns
cat_cols = ['IncidentGroup', 'FirstPumpArriving_DeployedFromStation']
data = pd.get_dummies(data, columns=cat_cols, prefix=cat_cols, prefix_sep='_')

# convert DateTimeOfCall to time of day
data['TimeOfDay'] = data['DatetimeOfCall'].dt.hour

# drop original datetime column
data.drop(columns=['DatetimeOfCall'], inplace=True)

# split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns=['FirstPumpArriving_AttendanceTime']), data['FirstPumpArriving_AttendanceTime'], test_size=0.2, random_state=42)

# fit linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# make predictions on test set
y_pred = model.predict(X_test)

# evaluate model performance
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("RMSE: {:.2f}".format(rmse))
print("R^2 Score: {:.2f}".format(r2))
```

```python
from sklearn.preprocessing import MinMaxScaler
# Convert DatetimeOfCall to datetime format
df['DatetimeOfCall'] = pd.to_datetime(df['DatetimeOfCall'], format='%Y-%m-%d %H:%M:%S')

# Extract hour, day of week, and month from DatetimeOfCall
df['Hour'] = df['DatetimeOfCall'].dt.hour
df['DayOfWeek'] = df['DatetimeOfCall'].dt.dayofweek
df['Month'] = df['DatetimeOfCall'].dt.month


# Preprocess the data
le = LabelEncoder()
df['SpecialServiceType'] = le.fit_transform(df['SpecialServiceType'])
df['StopCodeDescription'] = le.fit_transform(df['StopCodeDescription'])
# df['DatetimeOfCall'] = pd.to_datetime(df['DatetimeOfCall'])
df['DatetimeOfCall'] = df['DatetimeOfCall'].astype(int)

# Select the relevant columns
# X = df[['DatetimeOfCall', 'IncidentGroup']]
# X = df[['HourOfCall','IncidentGroup','SpecialServiceType','IncidentStationGround','NumCalls']]

# Scale Hour, DayOfWeek, and Month using MinMaxScaler
scaler = MinMaxScaler()
df[['Hour', 'DayOfWeek', 'Month']] = scaler.fit_transform(df[['Hour', 'DayOfWeek', 'Month']])

# Select the relevant columns
X = df[['Hour', 'DayOfWeek', 'Month', 'SpecialServiceType']]
y = df['StopCodeDescription'] # replace with the name of your target variable column

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Instantiate KNN with 4 neighbors
knn = KNeighborsClassifier(n_neighbors=4)

# Fit the model to the training data
knn.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = knn.predict(X_test)

# Calculate accuracy score and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
conf_mat = confusion_matrix(y_test, y_pred)

print('Accuracy:', accuracy)
print('Confusion Matrix:')
print(conf_mat)

sns.heatmap(conf_mat, annot=True, fmt='g', cmap='Blues')

# Set the axis labels and title
plt.xlabel('Predicted label')
plt.ylabel('True label')
plt.title('Confusion Matrix')
```

```python
from sklearn.metrics import classification_report
from tabulate import tabulate
# Calculate classification report
report = classification_report(y_test, y_pred, output_dict=True)

# Convert report to DataFrame and transpose it for easier viewing
df_report = pd.DataFrame(report).transpose()

# Rename the columns
df_report = df_report.rename(columns={'precision': 'Precision',
                                      'recall': 'Recall',
                                      'f1-score': 'F1 Score',
                                      'support': 'Support'})

# Print the report in a beautified manner
print("\nClassification Report:")
print(df_report.to_string(formatters={
    'Precision': '{:.2%}'.format,
    'Recall': '{:.2%}'.format,
    'F1 Score': '{:.2%}'.format,
    'Support': '{:,.0f}'.format
}))
print(tabulate(df_report, headers='keys', tablefmt='fancy_grid', floatfmt=".2f"))
```

```python
cols = ['Postcode_district', 'IncGeo_WardCode', 'IncidentGroup', 'PropertyCategory', 'PropertyType', 'StopCodeDescription']
df = df[cols]

# Remove rows with missing values
# df = df.dropna()

# Convert categorical columns to string type
df = df.astype(str)

# Apply one-hot encoding to the dataset
df_onehot = pd.get_dummies(df)

# Apply apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df_onehot, min_support=0.05, use_colnames=True)

# Apply association rule mining to find interesting rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Print the top 10 rules by lift value
print(rules.nlargest(10, 'lift'))
```

```python
conf_rules = rules[(rules['confidence'] > 0.5) & (rules['lift'] > 1.5)]
num_rules_arm = len(conf_rules)
mean_conf_arm = conf_rules['confidence'].mean()
print(mean_conf_arm)
```
